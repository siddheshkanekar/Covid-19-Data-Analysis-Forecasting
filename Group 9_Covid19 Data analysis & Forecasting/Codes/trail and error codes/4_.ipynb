```python
import pandas as pd
import numpy as np
```

```python
x = np.array([0.2,1.2,1,1.4,-1.5,0.5,-0.5])
y = x*3+5
print(x)
print(y)
```

```python
i_w = 0.5
i_b = -0.5
eta = 0.2
ypred = i_b+i_w*x
print(ypred)
```

```python
Loss = np.sum((y-ypred)**2)*(1/2*len(x))
Loss
```

```python
db = -(1/len(x))*np.sum(y-ypred)
dw = -(1/len(x))*np.sum((y-ypred)*x)
```

```python
w,b = i_w,i_b
new_w = w-eta*dw
new_b = b-eta*db
print(new_w,new_b)
```

```python
i_w = 0.5
i_b = -0.5
eta = 0.2
w,b = i_w,i_b
for epoch in range(1,101):
    ypred = b + w * x
    Loss = np.sum((y - ypred) ** 2) * (1/2 * len(x))
    print("loss: ",Loss)
    if Loss<0.0001:
        break
    db = -(1 / len(x)) * np.sum(y - ypred)
    dw = -(1 / len(x)) * np.sum((y - ypred) * x)
    new_w = w - eta * dw
    new_b = b - eta * db
    print(new_w, new_b)
    w,b = new_w,new_b
```
